```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(f1, f2, t0, tn, x0, y0, h):
    n=int((tn-t0)/h)
    t=np.linspace(t0, tn, n+1)
    x=np.zeros(n+1)
    y=np.zeros(n+1)
    x[0]=x0
    y[0]=y0
    for i in range(1, n+1):
        x[i]=x[i-1]+h*f1(t[i-1], x[i-1], y[i-1])
        y[i]=y[i-1]+h*f2(t[i-1], x[i-1], y[i-1])
        if x[i] < 0:
            # bounce!
            x[i] = 0
            y[i] = -y[i] * 0.8
    return((t ,x, y))

```

```python
def g1(t,x,y):
    return(y)

def g2(t, x, y):
    return(-10)

(t1, x1, y1)=euler(g1, g2, 0, 3, 5, 0, 0.0001)

plt.plot(t1, x1, label="h-0.0001")
plt.legend()

```

```python
def euler2(f1, f2, f3, f4, t0, tn, x0, y0, z0, w0, h):
    n=int((tn-t0)/h)
    t=np.linspace(t0, tn, n+1)
    x=np.zeros(n+1)
    y=np.zeros(n+1)
    z=np.zeros(n+1)
    w=np.zeros(n+1)
    x[0]=x0
    y[0]=y0
    z[0]=z0
    w[0]=w0
    for i in range(1, n+1):
        x[i]=x[i-1]+h*f1(t[i-1], x[i-1], y[i-1], z[i-1], w[i-1])
        y[i]=y[i-1]+h*f2(t[i-1], x[i-1], y[i-1],  z[i-1], w[i-1])
        z[i]=z[i-1]+h*f3(t[i-1], x[i-1], y[i-1],  z[i-1], w[i-1])
        w[i]=w[i-1]+h*f4(t[i-1], x[i-1], y[i-1],  z[i-1], w[i-1])
        if x[i] < 0:
            # bounce!
            x[i] = 0
            y[i] = -y[i] * 0.8
        if z[i]> 3:
           z[i]=3
           w[i]=-w[i]* 0.8
        if z[i]<0:
           z[i]=0
           w[i]=-w[i]* 0.8
    return((t , x, y, z, w))
```

```python
def g1(t,x,y,z, w):
  return(y)

def g2(t, x, y, z, w):
  return(-10)

def g3(t, x, y, z, w):
  return(w)

def g4(t, x, y, z, w):
  return(0)


(t1, x1, y1, z1, w1)=euler2(g1, g2, g3, g4, 0, 10, 5, 0, 0, 1, 0.0001)

```

```python
plt.plot(t1, x1, label="h-0.0001")
plt.legend()
```

```python
plt.plot(t1, z1, label="h-0.0001")
plt.legend()
```

```python
plt.plot(z1, x1, label="h-0.0001")
plt.legend()
```

```python
def h1(t,x,y,z, w):
  return(y)

def h2(t, x, y, z, w):
  return(-10-y)

def h3(t, x, y, z, w):
  return(w)

def h4(t, x, y, z, w):
  return(-w)


(t2, x2, y2, z2, w2)=euler2(h1, h2, h3, h4, 0, 10, 5, 0, 0, 1, 0.0001)
```
