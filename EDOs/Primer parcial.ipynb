```python
def planetaRepulsivo(X, t, G, m, r=3):
    x1, x2, v1, v2 = X    # unpack variables
    dx1 = v1
    dx2 = v2
    dv1 = (G * x1)/(np.sqrt(x1**2+x2**2))**r
    dv2 = (G * x2)/(np.sqrt(x1**2+x2**2))**r
    dXdt = [dx1, dx2, dv1, dv2]    # pack derivatives
    return dXdt
```

```python
m=1
G=1

r=1 #probar distintos r


x1i, x2i = 1, 1
v1i, v2i = 0.5, 0.5
init = [x1i, x2i, v1i, v2i]

T = 20.
time = np.arange(0, T, 0.1)

sol = integrate.odeint(planetaRepulsivo, init, time, args=(m, G, r))
X = sol[:,0:2]
```

```python
plt.plot(sol[:,0], sol[:,1])
plt.legend()
```

```python
a = 1
q = 1
H0 = 10

r=0.2

def F(h):
  return 1 - math.sqrt(h)
```

```python
def euler(t0,tf,y0,h,f,l):
    t = np.arange(t0,tf+h,h)
    y = [y0]
    for i in range(len(t)-1):
        y.append(y[-1]+h*f(t[i],y[-1],l))
    return t,y
```

```python
def f_exp(t,y,r):
    return 1-(np.sqrt(y) / r)
```

```python
sol_h1 = euler(0,2,1,0.1,f_exp,0.1)
```

```python
plt.plot(sol_h1[0], sol_h1[1], c = 'r')
```

```python
rs = [0.1,0.2,0.5,1,2,np.sqrt(10),5]
xs = []
i=0

for h in rs:
    x, y = euler(0,2,1,0.1,f_exp,h)
    plt.plot(x, y, label=f'R = {h}')
    i= i+1

t_exact = np.linspace(0, 2, 100)


plt.xlabel('t')
plt.ylabel('h')
plt.title('Euler Method Solutions for Different R')
plt.legend()
plt.show()
```
