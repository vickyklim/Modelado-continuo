```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[12, 12]
plt.rcParams.update({'font.size':18})
dt=0.001
t=np.arange(0,1, dt)
f=np.sin(2*np.pi*12*t)+np.sin(2*np.pi*5*t)
g=f+2.5*(np.random.rand(len(t))-0.5)
plt.plot(f)
plt.plot(g)
```

```python
n=len(t)
fhat=np.fft.fft(g,n)
PS=fhat*np.conj(fhat)/n
L=np.arange(1, np.floor(n), dtype='int_')
plt.plot(PS[L])
```

Tiene pequeo ruidito


```python
indices=PS>100
PSlimpio=PS*indices
fhat=indices*fhat
ffclean=np.fft.ifft(fhat)
plt.plot(ffclean)
plt.plot(f)
```
