```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y(t, f):
  return np.sin(2*np.pi*f*t)

N=20
f=1
dt=0.2
fn=1/(2*dt)
df=1/(2*N*dt)
tsampled=np.arange(0, N*dt, dt)
tsmooth=np.arange(0, N*dt, dt/100)
ysampled=y(f, tsampled)
ysmooth=y(f, tsmooth)
plt.plot(tsmooth, ysmooth)
plt.plot(tsampled, ysampled, "o")
```

```python
c=np.fft.fft(ysampled)
freq=np.arange(0, 2*fn, 2*df)
plt.plot(freq, abs(c))
```
