```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0=1

def h(x, x0):
  y=0.5*(x-x0+np.abs(x-x0))/(x-x0)
  return(y)
x=np.linspace(-4,4, 1000)
plt.plot(x, h(x, x0))
```

```python
#Escalon entre uno y tres#

plt.plot(x, h(x, 1)-h(x,3))
```

```python
#Funci√≥n pozo#

plt.plot(x, -h(x, 1)+h(x,3))
```

```python
#Funci√≥n pozo levantada en 1#

plt.plot(x, -h(x, 1)+h(x,3)+1)
```
