```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[12, 12]
plt.rcParams.update({'font.size':18})
dt=0.001
#N = 1/delta = 1000
#freq sampleo = 1/delta =1000
#Nyquist = 1000/2 = 500
t=np.arange(0,1, dt)
f=np.sin(2*np.pi*12*t)+np.sin(2*np.pi*5*t)
g=f+2.5*np.random.rand(len(t))
plt.plot(f)
plt.plot(g)
```

```python
n=len(t)
fhat=np.fft.fft(g,n)
PS=fhat*np.conj(fhat)/n
L=np.arange(1, np.floor(n/2), dtype='int_')
plt.xlim=(0,15)
plt.plot(PS[L])



```

```python
def DFT(x):
  N=len(x)
  X=[]
  for k in range(N):
    X_k=0
    for n in range(N):
      e=np.exp((2j*np.pi*k*n)/N)
      X_k+=x[n]/e
    X.append(X_k)
  return np.array(X)

plt.plot(np.abs(DFT(f)))
plt.xlim=(0,15)
```
